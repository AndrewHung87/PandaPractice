* Group together rows that fall under the same category.
* Perform an operation that aggregates across all rows in the category.

# Skew
* A distribution with a long tail **right** is skewed right.
* A distribution with a long tail **left** is skewed left.

#Displot

`displot` is a wrapper for histplot, kdeplot, and ecdfplot to plot distribution.

* histplot: `sns.displot(data=df, kind="hist")`

* kdeplot: `sns.displot(data=df, kind="kde")`

* ecdfplt: `sns.displot(data=df, kind="ecdf")`
