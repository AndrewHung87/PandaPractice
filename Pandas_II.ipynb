```python
import glob
```

```python
file_path = "/content/*.TXT" # Define the path where the TXT files are located

all_files = glob.glob(file_path) # Get a list of all state TXT files

column_names = ["State", "Gender", "Year", "Name", "Count"] # Define column names based on SSA format

dfs = []

# Read all TXT files and append to the list
for file in all_files:
    df = pd.read_csv(file, header=None, names=column_names)
    dfs.append(df)

# Merge all dataframes into one cumulative dataframe
cumulative_df = pd.concat(dfs, ignore_index=True)

# Display the first few rows
print(cumulative_df.head())

# Save as CSV
cumulative_df.to_csv("/content/babynames.csv", index=False)

# Download the file
from google.colab import files
files.download("/content/babynames.csv")
```

```python
def dr_ea_count(string):
  return string.count('dr') + string.count('ea')

# Use map to apply dr_ea_count to each nmame in the "Name" column
babynames["dr_ea_count"] = babynames["Name"].map(dr_ea_count)
babynames = babynames.sort_values("dr_ea_count", ascending=False)
babynames.head()
```
